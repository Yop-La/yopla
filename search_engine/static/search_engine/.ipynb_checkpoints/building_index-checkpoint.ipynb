# Construction de l'index

Dans ce notebook, il s'agit de construire un index de recherche à partir d'un corpus de 2500 textes. Nous allons d'abord charger en mémoire les textes, puis tokeniser les textes et enfin construite l'index. Toutes les étapes sont détaillés avec des commentaires

L'index construit contiendra également des métriques qui seront utilisés pour trier les résultats par ordre de pertinence par la suite

```python
from __future__ import print_function
import os, sys
import string
import re
import nltk
import string
#librairie pour faire du steeming ( réduire les mots à leurs racines) 
from nltk.stem.porter import PorterStemmer
import collections;
import time
from collections import defaultdict
import numpy as np
from pickle import dump, load
# pour récupérer les stopwords dans la liste sw
nltk.download('stopwords')
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize 
import itertools
import re
from math import log
from functools import reduce
from multiprocessing import Pool
import dill
import pickle
```

```python
# charge les 2500 textes et les enregistre dans list_text

path = './data'
list_text = []
        
files = os.listdir(path)
for name in files:
    path = './data/' + name
    files_txt = os.listdir(path)
        
    for name_txt in files_txt:    
        path_txt = './data/' + name + '/' +name_txt
        with open(path_txt, 'r') as myfile:
            data = myfile.read()
        list_text.append(data)

D = len(list_text)
print("Il y a " + str(len(list_text)) + " textes et le premier texte est: ");
print(list_text[0]);
```

```python
with open('list_text.pkl', 'wb') as f:
    dump(list_text,f)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
%%time
#Cette cellule retourne une liste de tokens nomée list_tokens
# ie : list_tokens[1] est une liste des tokens (des mots normalisés en minuscule, sans stop word et ponctuations) 
# du texte 2

list_tokens = []
ps = PorterStemmer()

#fonction tokenise qui à partir d'un texte retourne une liste de token normalisé
def tokenise( text ):
    # split into words
    tokens = word_tokenize(text)
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove punctuation from each word
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens] 
    # stemming of words
    stemmed = [ps.stem(word) for word in tokens]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words
    words = [w for w in words if not w in stop_words]
    return(words)

list_tokens = list(map(tokenise, list_text))

print(len(list_tokens))
print(list_tokens[0])
```

```python
%%time

words = itertools.chain( * list_tokens )
words = list(words)
print("words")
print(len(words))
#retourne tous les mots normalisés du corpus de texte
```

```python
%%time
#construction du vocabulaire ( ie tous les mots du texte normalisés apparaissant plus de 10 fois)
# on construit d'abord un dictionnaire words_count ( clé : un mot ; valeur : nb d'occurence du mots dans le corpus)
words_count =  collections.Counter(words)

d = dict((k, v) for k, v in words_count.items() if v >= 10)

vocs = set(list(words_count.keys()))#vocabulaire 
print("taille du voc")
print(len(vocs))
```

```python
# les 10 mots les plus communs
words_count.most_common(10)
```

```python
# les 10 mots les moins communs
words_count.most_common()[-10:]
```

```python
with open('voc.pkl', 'wb') as f:
    dump(vocs,f)
```

```python
%%time

#construction de l'index : une clé = un mot du voc et la valeur = une liste de dictionnaire
# où un dictionnaire contient : 
#    -en premier élément : l'indice du texte contenant le mot dans le corpus
#    -en deuxième élément : la liste des positions du mot dans le texte

index = defaultdict(list)
i=0
for list_token in list_tokens:    
    arr_token = np.array(list_token)
    set_token = set(list_token)
    set_token = set_token.intersection(vocs) # ne contient que les tokens du voc
    d_len = len(set_token)

    
    
    for word in set_token:
        list_pos = list(np.where(arr_token == word)[0])
        ndt=len(list_pos)
        
        tf = ndt/d_len
        dic = {'indice_text':i,'pos_in_text':list_pos,'ndt':ndt,'tf':tf}
        index[word].append(dic)
        

    i=i+1

# pour calculer tf_idf
for word in index.keys():
    res_index = index[word]
    for res in res_index:  
        nt = len(res_index)
        idf = log(D/(1+nt))
        tf_idf = idf*res['tf']
        res['tf_idf'] = tf_idf
        res['nt'] = nt
```

```python
%%time
# construction de l'index avec map/reduce

def concat_dict(x,y):
     return {**x, **y}
    
def build_index(a_tuple):
    i = a_tuple[0]
    list_token = a_tuple[1]
    index = defaultdict(list)
    arr_token = np.array(list_token)
    set_token = set(list_token)
    set_token = set_token.intersection(vocs) # ne contient que les tokens du voc
    for word in set_token:
        list_pos = list(np.where(arr_token == word)[0])
        ndt=len(list_pos)
        d_len = len(set_token)
        tf = ndt/d_len
         
        dic = {'indice_text':i,'pos_in_text':list_pos,'ndt':ndt,'tf':tf}
        index[word].append(dic)
    return(index)

#l est list_okens
#f est une fonction à définir ( l'intérieur de la boucle)
    
with Pool(5) as pool:
     sq = pool.map(build_index,enumerate(list_tokens),200)


reduce (lambda a,b: concat_dict(a,b), sq)


# De moins bonnes performances avec map - reduce. Envisager d'adopter un autre algo pour faire ce map-reduce
```
